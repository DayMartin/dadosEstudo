#### Indexação com NumPy

```python
print(arr1)
```

```python
arr1[4]
```

```python
#### trzer os números entre a posição 1 ao 4
### Lembrando que o último valor solicitado é exclusivo, então solicitamos de 1 a 4
### mas ele vai imprimir de 1 a 3
arr1[1:4]
```

```python
### Indexação trzer os números entre a posição 1 ao 4 + 1
### Já aqui por estarmos somando, eu to pedindo para ele trazer de 1 a 4 e mais 1
### sabemos que de 1 a 4 ele traria somente 3 porque o último número é reservado
### então ele traz vai trazer de 1 a 3 + 1 
arr1[1:4+1]
```

```python
#### seria o mesmo que fazer isto, ele vai reservar o valor do indice 5, e vai me trazer
#### os 4 valores
arr1[1:5]
```

```python
### criando uma lista de indices
indice = [1,2,5,6]
```

```python
### imprimindo os indices baseado na lista de indices
arr1[indice]
```

```python
#### criar uma mascara (um padrão)
#### Gravar na variavel mask todos os valores que dividido por 2 tenha o resto 0 ( par )
#### Vai percorrer todos os elementos de arra1 e realizar a mascara
mask = (arr1 % 2 == 0)
```

```python
mask
```

```python
### Agora vou utilizar a mask como indice, todos os valores que tiverem o valor boolean True
arr1[mask]
```

```python
#### Valor antes da alteração abaixo
arr1[0]
```

```python
#### alterar um elemento do array
arr1[0] = 1
```

```python
#### Valor depois da alteração acima
print(arr1[0])
```

```python
#### Na lista python é possível atribuir valores de outros tipos, por exemplo
#### É possivel acrescentar na lista numeros e string
#### Já no NumPy array, não é possível fazer isso
#### Se você criou uma array de números, vocẽ não pode acrescentar uma string

try: 
    arr1[0] = 'Teste'
except:
    print("Não é possível")
```
