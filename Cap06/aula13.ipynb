#### Reduce
Utilizado para reduzir em um único valor uma lista de elementos

```python
from functools import reduce
```

```python
#### Veja nesta imagem, eu tenho uma lista de 4 números 
#### O resultado da soma entre o primeiro e segundo número, eu quero somar com o terceiro número
#### O resultado da soma entre resultado do primeiro e seundo numero com o numero 3 eu vou somar com o 4º número
#### Esse é o papel do reduce

from IPython.display import Image
Image('arquivos/reduzir.png')
```

```python
numeros = [47, 11, 42, 13]
```

```python
def somar(a,b):
    x = a+b
    return x
```

```python
reduce(somar, numeros)
```

Usando a função Reduce com lambda

```python
#### já temos a lista 
lts = [47, 11, 42, 13]
```

```python
### Utilizamos lambda
reduce(lambda a,b: a+b, lts)
```

Agora vamos retornar o maior valor da lista

```python
#### Atribuimos a função lambda à uma variavel
list_max = lambda a,b: a if(a > b) else b
```

```python
reduce(list_max, lts)
```
